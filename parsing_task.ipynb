```python
url = 'https://easychair.org/cfp/' 
import re
import urllib2, json, sys, time
res  = urllib2.urlopen(url).read()
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res, 'html.parser')
tag=soup.find_all('tbody')
imp=tag[1]
first_look_table=imp.find_all('td')
```

```python
first_look_table=imp.find_all('td')
```

```python
data=[[],[],[],[],[],[],[],[],[],[],[]]
hrefs=[]
hrefs.append(first_look_table[0].a.get('href'))
data[0].append(first_look_table[0].string)
data[1].append(first_look_table[1].string)
data[2].append(first_look_table[2].string)
data[3].append(first_look_table[3].string)
data[4].append(first_look_table[4].string)
data[5].append([])
for i in first_look_table[5]('a'):
    data[5][0].append(i.string)
```

```python
for i in hrefs: 
    res  = urllib2.urlopen(i).read()
    soup = BeautifulSoup(res, 'html.parser')
    [s.extract() for s in soup(['style', 'script', '[document]', 'head', 'title'])]
    visible_text = soup.getText()
    try:
        data[7].append(re.findall(r'Conference website(.*?)Submission',visible_text)[0])
    except:
        data[7].append(i)
    z=re.findall(r'<div id="cfpdateplace">(.*?)</div>',str(soup.find_all('div',id="cfpdateplace")))[0].split(',')
    if '-' in z[len(z)-2]:
        z2=z[len(z)-2].split('-')
        if len(z2[1])<=2:
            data[8].append(z2[0].strip())
            data[9].append(z2[0][:len(z2[0])-3].strip()+" "+z2[1].strip())
        else:
            data[8].append(z2[0].strip())
            data[9].append(z2[1].strip())
    else:
        data[8].append(z[len(z)-2].strip())
        data[9].append(z[len(z)-2].strip())
    data[10].append(z[len(z)-1].strip())
    norm_text=list()
    for j in soup.find_all('p'):
        norm_text.append(j.get_text())
    fnt=str()
    for j in norm_text:
        fnt+=j+'\n '
    data[6].append(fnt)
```

```python
js={}
for i in range(172):
    js[data[0][i]]={}
    js[data[0][i]][u'category']=""
    js[data[0][i]][u'startDate']=data[8][i]
    js[data[0][i]][u'final version due']=data[3][i]
    js[data[0][i]][u'endDate']=data[9][i]
    js[data[0][i]][u'publisher']=""
    js[data[0][i]][u'url']=data[7][i]
    js[data[0][i]][u'text']=data[6][i]
    js[data[0][i]][u'topics']=data[5][i]
    js[data[0][i]][u'year']=data[10][i]
    js[data[0][i]][ u'url_cfp']=hrefs[i]
    js[data[0][i]][u'submission deadline']=data[3][i]
    js[data[0][i]][u'location']=data[2][i]
    js[data[0][i]][u'full title']=data[1][i]
    js[data[0][i]][u'demo_paper_length']=''
    js[data[0][i]][u'short_paper_length']=''
    js[data[0][i]][u'short title']=data[0][i]
    js[data[0][i]][u'type']=""
    js[data[0][i]][u'id']=i
    js[data[0][i]][u'full_paper_length']=''
    js[data[0][i]][u'notification due']=data[3][i]
```
